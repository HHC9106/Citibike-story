```python
import requests
import numpy as np
import pandas as pd
import datetime
import time
```

```python
url = 'https://gbfs.citibikenyc.com/gbfs/en/station_information.json'
r = requests.get(url)
data = r.json()
station_info_df = pd.DataFrame(data['data']['stations'])
```

```python
station_link = 'https://gbfs.citibikenyc.com/gbfs/en/station_status.json'
station_info_link = 'https://gbfs.citibikenyc.com/gbfs/en/station_information.json'

timeout = time.time() + 60*5   # 5 minutes from now

def get_station_df(station_info_df, station_status_link):
    
    with requests.get(station_status_link) as url:
        data = url.json()
        station_status_df = pd.DataFrame(data['data']['stations']).set_index('station_id')
    
    station_df = pd.merge(station_status_df, 
                      station_info_df, 
                      left_index=True, 
                      right_index=True)
    station_df = station_df.assign(ts=pd.Series([datetime.datetime.now()] * len(station_df)).values)
    return station_df
    

with requests.get(station_info_link) as url:
    data = url.json()
    station_info_df = pd.DataFrame(data['data']['stations']).set_index('station_id')    
    
while True:
    station_df = get_station_df(station_info_df, station_link)
    print(station_df)
    station_df.to_csv('../data/station_status.csv', mode='a', header=False)
    time.sleep(60)

    #Set a stop time
    test = 0
    if test == 5 or time.time() > timeout:
        break
    test = test - 1
```
